```python
from mlflow.tracking import MlflowClient

# Step 1: Create an MlflowClient instance
client = MlflowClient(tracking_uri="file:///home/jlagesse/ngmb/mlruns")

# Step 2: Retrieve all experiments
experiments = client.search_experiments(filter_string="name LIKE '%small'", max_results=1000)

experiment_ids = []
# Step 3: Print the experiment details
for experiment in experiments:
    print(f"Name: {experiment.name}, Artifact Location: {experiment.artifact_location}")
    experiment_ids.append(experiment.experiment_id)
```

```python
all_runs = client.search_runs(experiment_ids=experiment_ids, max_results=1000)
```
